## VGGNet

```python
from tensorflow.keras.applications.vgg16 import VGG16, decode_predictions
```

```python
vgg = VGG16()
vgg.summary()
```

### Preprocessing

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
img = np.array(Image.open('images/cat.jpg').resize((224, 224)))
plt.imshow(img)
plt.axis('off')
plt.show()
```

### Processing

#### Predicting

```python
yhat = vgg.predict(img.reshape(-1, 224, 224, 3))
label = decode_predictions(yhat)
label = label[0][0]
print(f'{label[1]} ({label[2] * 100:.2f}%)')
```

```python
img = np.array(Image.open('images/dog.jpg').resize((224, 224)))
yhat = vgg.predict(img.reshape(-1, 224, 224, 3))
label = decode_predictions(yhat)
label = label[0][0]
print(f'{label[1]} ({label[2] * 100:.2f}%)')
```
