#### 2. Quelle langue

```python
import requests
from urllib.parse import quote
```

```python
with open('kakao api.txt', mode = 'r') as key_file :
    kakao = key_file.read()
```

```python
def speak(text) :
    base_url = 'https://dapi.kakao.com/v3/translation/language/detect'
    url = f'https://dapi.kakao.com/v3/translation/language/detect?query={quote(text)}'
    result = requests.get(url, headers = {'Authorization' : f'KakaoAK {kakao}'}).json()
    print(f'{result["language_info"][0]["name"]}, {round(result["language_info"][0]["confidence"] * 100)}%')
```

```python
speak("Bonjour, Je m'appelle Eugene. Enchat!")
```
