#### 2. Quelle langue et comment on dit en coréen?

```python
import requests
from urllib.parse import quote
```

```python
with open('kakao api.txt', mode = 'r') as key_file :
    kakao = key_file.read()
```

```python
def translator(text) :
    url = f'https://dapi.kakao.com/v3/translation/language/detect?query={quote(text)}'
    result = requests.get(url, headers = {'Authorization' : f'KakaoAK {kakao}'}).json()
    lang = result["language_info"][0]["name"]
    per = round(result["language_info"][0]["confidence"] * 100)
    code = str(result["language_info"][0]["code"])

    url = f'https://dapi.kakao.com/v2/translation/translate?src_lang={code}&target_lang=kr&query={quote(text)}'
    result = requests.get(url, headers = {'Authorization' : f'KakaoAK {kakao}'}).json()
    output = '\n'.join([i[0] for i in result['translated_text']])
    print(f'Korean : {output}')
    print(f'Language : {lang}')
    print(f'Accuracy : {per}%')
```

```python
translator("""
Bonjour à tous!
je m'appelle Eugene.
Enchanté!
""")
```
