```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset/ThoraricSurgery.csv')
df
```

### Creating dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df.iloc[:, :-1].values, df.iloc[:, -1].values,
    stratify = df.iloc[:, -1].values, random_state = 2021
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### Defining model

```python
import tensorflow as tf
seed = 2021
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(30, input_shape = (17,), activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
model.summary()
```

### Model compiling - 실행 환경 설정

```python
model.compile(
    loss = 'binary_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)
```

모델 학습

```python
model.fit(
    X_train, y_train,
    validation_split = 0.2,
    epochs = 30,
    batch_size = 30
)
```

### Accuracy evaluation

```python
model.evaluate(X_test, y_test)
```
